```python
import socket
import threading
import time
 
ENCODING = 'utf-8'
  
class ab(threading.Thread):
 
    def __init__(self,r_port,table,my_port,i):
        threading.Thread.__init__(self)
        self.r_port =r_port
        self.table = table
        self.my_port=my_port
        self.i=i;
        
        self.leader=0;
 
    def listen(self):
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.bind(('127.0.0.1', self.my_port))
        sock.listen(10)
        try:
            
            
            while True:
            
            
                connection, client_address = sock.accept()
                
                while True:
                
                    data1 = connection.recv(1024);
                
                
                    if data1 :
                        mm=message(**json.loads(data1,encoding='utf-8'));
                    
                        self.table[self.i][2]=mm.value
                        self.table[self.i][4]=mm.type
                    
                        
                    
                    if not data1:
                        print('node '+str(self.table[self.i][0])+': '+str(mm.type)+' '+str(mm.value)+'\n')
                        break;
                    
                if self.table[self.i][5]=='1':
                
                    break;
        finally:
            
            sys.exit();
            
            
                
                    
 
    
    
    
    def send(self):
        flag=1
        try:
            
            while True:
            
            
                if flag==0 and self.table[self.i][2]<self.table[self.i][0] and self.table[self.i][3]==0:
                
                
                    self.table[self.i][0]=self.table[self.i][2]
                    time.sleep(int(self.table[self.i][1]))
                
                    m = message('ELECTION',self.table[self.i][2])
                    mm=json.dumps(m.__dict__).encode("utf-8")
                
                    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
                    s.connect(('127.0.0.1', self.r_port))
                    s.sendall(mm)
                    s.close()
            
            
            
            
            
            
            
            
            
                if flag==0 and self.table[self.i][2]>self.table[self.i][0] and self.table[self.i][4]=='ELECTION':
                
                
                
                    time.sleep(int(self.table[self.i][1]))
                    m = message('ELECTION',self.table[self.i][2])
                    mm=json.dumps(m.__dict__).encode("utf-8")
                    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
                    s.connect(('127.0.0.1', self.r_port))
                    s.sendall(mm)
                
                    self.table[self.i][2]=0;
                    self.table[self.i][3]=1;
                    s.close()
            
            
            
            
                if flag==0 and self.table[self.i][0]!=self.table[self.i][2] and self.table[self.i][4]=='LEADER':
                
                
                    time.sleep(int(self.table[self.i][1]))
                    m = message('LEADER',self.table[self.i][2])
                    mm=json.dumps(m.__dict__).encode("utf-8")
                    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
                    s.connect(('127.0.0.1', self.r_port))
                    s.sendall(mm)
                    s.close()
                    self.table[self.i][3]=0
                    self.table[self.i][5]=1;
                    break;
                
                
                
            
          
                
            
            
                if flag==0 and self.table[self.i][2]==self.table[self.i][0] and self.table[self.i][4]=='ELECTION':
                
                    time.sleep(int(self.table[self.i][1]))
                    m = message('LEADER',self.table[self.i][0])
                    mm=json.dumps(m.__dict__).encode("utf-8")
                    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
                    s.connect(('127.0.0.1', self.r_port))
                    s.sendall(mm)
                
                    self.table[self.i][5]=1;
                    self.table[self.i][2]=0;
                    self.leader=self.table[self.i][0];
                    s.close();
                    self.table[self.i][3]=0;
                
            
            
                if flag==0 and self.table[self.i][2]==self.table[self.i][0] and self.table[self.i][4]=='LEADER':
                    self.table[self.i][5]=1;
                
                    break;
                
            
            
                if flag==1:
                
                    flag=0
                    time.sleep(int(self.table[self.i][1]))
                    m = message('ELECTION',self.table[self.i][0])
                    mm=json.dumps(m.__dict__).encode("utf-8")
                    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
                    s.connect(('127.0.0.1', self.r_port))
                    s.sendall(mm)
                    self.table[self.i][3]=1;
                    s.close()
                
        
        finally:
            sys.exit();
    
            
    
    def run(self):
        
        for
        
        
        t1=threading.Thread(target=self.listen)
        t2=threading.Thread(target=self.send)
        
        t1.start()
        t2.start()
        
        t1.join()
        t2.join()
        if self.table[self.i][0]==self.leader:
            
            time.sleep(int(self.table[self.i][1]))
            os._exit(0);
            
            
        
        
    
    
    
    
```

```python
def LCR(n,a):
    threads=[];
    for j in range(n):
        if j==n-1:
            p=0
        else:
            p=j+1;
        
        s=ab(9090+int(a[p][0]),a,9090+int(a[j][0]),j);
        t=threading.Thread(target=s.run)
        t.start()
        threads.append(t)
    for thread in threads:
        thread.join()        
LCR(n,a)        
```
